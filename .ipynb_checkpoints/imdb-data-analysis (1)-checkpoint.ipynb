*Spelling correction*

```python
from textblob import TextBlob
def spelling_corr(data):
    return TextBlob(data).correct().string

df['review'].apply(spelling_corr)
```

*Punctuations*

```python
import string

string.punctuation
def remove_punctuations(data):
    for i in string.punctuation:
        if i in data:
            data = data.replace(i,'')
    return data

df['review'] = df['review'].apply(remove_punctuations)
```

**PREPROCESSING**

*Tokeniation*

```python
from nltk.tokenize import word_tokenize

df['tokenized review'] = df['review'].apply(word_tokenize)
```

```python
df.head()
```

*Stop word removal*

```python
from nltk.corpus import stopwords

L = []
def remove_stopwords(data):
    for word in data:
        if word not in stopwords.words('english'):
            L.append(word)
    return L

df['tokenized review'] = df['tokenized review'].apply(remove_stopwords)
```

```python
df.head()
```

```python
#updating review column

df['review'] = df['tokenized review'].apply(lambda x: " ".join(x))
```
